```python
#範例1
import json
import csv
file = open('GE1-1-input.json','r',encoding='utf8')
data = json.load(file)

file2 = open('GE1-1-output.csv','w',encoding='utf8',newline='')
w = csv.writer(file2)

for row in data:
    w.writerows([[row['sno'],row['sna'],row['tot']]])
    
file.close()
file2.close()
```

```python
# 範例2
import json
import csv
import requests
url="https://data.coa.gov.tw/Service/OpenData/FromM/FarmTransData.aspx?$top=1000&$skip=0"
res = requests.get(url)
data = json.loads(res.text)
# print(data)
file = open('GE1-2-output.csv','w',encoding='utf8',newline='')
w = csv.writer(file)
for row in data:
    F0 = row['交易日期']
    F1 = row['作物名稱']
    F2 = row['市場名稱']
    F3 = row['平均價']
    F4 = row['交易量']
    w.writerows([[F0,F1,F2,F3,F4]])

file.close()
```

```python
#範例3
import json
from urllib.request import urlopen
import gzip

url = "https://tcgbusfs.blob.core.windows.net/blobyoubike/YouBikeTP.gz"
html = urlopen(url)
file = gzip.GzipFile(fileobj = html)
jdata = file.read()
file.close()

data = json.loads(jdata)
# print(data['retVal'].items())

file = open('GE1-3-output.txt','w',encoding='utf8')
for key,value in data['retVal'].items():
    file.write('NO.' + value['sno']+" "+value['sna']+" "+value['mday']+" "+value['sbi']+"\n")
    
file.close()
```

```python
#範例5
import xml.etree.ElementTree as et
import csv

tree = et.parse('GE1-5-input.xml')
root = tree.getroot()


file = open('GE1-5-output.csv','w',encoding='utf8',newline='')
w = csv.writer(file)

for row in root:
    sno = row.find('sno').text
    sna = row.find('sna').text
    tot = row.find('tot').text
    w.writerows([[sno,sna,tot]])

file.close()
```

```python
import csv
import xml.etree.ElementTree as et
tree = et.parse('GE1-7-input.xml')
root = tree.getroot()

file = open('GE1-7-output.csv','w',encoding='utf8',newline='')
w = csv.writer(file)

for row in root:
    C1 = row.find('縣市別').text
    C2 = row.find('醫療機構名稱').text
    C3 = row.find('醫療機構地址').text
    w.writerows([[C1,C2,C3]])

file.close()
```

```python
import sqlite3 as db
import csv

conn = db.connect('Supplier.db')

conn.execute("""

create table if not exists Supplier(
    Supplier_Name varchar(20),
    Part_Number varchar(20),
    Cost float,
    Purchase_Date date
)

""")
print("create table access")

file = open('data.csv','r',encoding='utf8')
r = csv.reader(file)

count = 0
for row in r :
    if count > 0 :
        conn.execute("""

        insert into Supplier(Supplier_Name,Part_Number,Cost,Purchase_Date)
        values(?,?,?,?)

        """,(row[0],row[1],row[2],row[3],))

    count +=1
    
conn.commit()
file.close()
print("insert data success")
    
data = conn.execute("""

select * from Supplier

""")

data = data.fetchall()

print(data)

print("----------------------------")

for row in data:
    print(row[0]+","+row[1]+","+row[2]+","+row[3]+",")


conn.close()    



```
