```python
import json
import csv

file = open('GE1-1-input.json','r',encoding='utf8')
data = json.load(file)

output_file = open('GE1-1-output.csv','w',encoding='utf8',newline='')
r = csv.writer(output_file)

for row in data:
    r.writerows([[row['sno'],row['sna'],row['tot']]])
    
file.close()
output_file.close()
```
